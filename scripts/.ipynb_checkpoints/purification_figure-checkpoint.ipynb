```python
# batch 5
import pandas as pd
fig,ax = plt.subplots(figsize=(10,10))
s200=pd.read_csv('./DAB2-Batch5/18Sept27_DABS200/18Sept27_DAB_S200Inrease_Cleaned.csv')
plt.plot(s200['UV 1_280_ml'],s200['UV 1_280_mAU'],label='UV 280')
plt.plot(s200['UV 2_485_ml'],s200['UV 2_485_mAU'],label='UV 485')
plt.xlim(0,4)
plt.ylabel('mAU',fontsize=15)
plt.xlabel('ml',fontsize=15)

peak=s200['UV 1_280_ml'][s200['UV 1_280_mAU']==max(s200['UV 1_280_mAU'])].iloc[0]
bX=1.15
bY=443
lX=1.27
lY=224
slope=(lY-bY)/(lX-bX)
kDa = bY+slope*(peak-bX)
plt.axvline(peak,label=kDa,c='g')

plt.axvline(bX,label=bY,c='r')
plt.axvline(lX,label=lY,c='k')

plt.legend()
```

```python
# making Panel C
%matplotlib inline
def makePanelC(ax):
    import pandas as pd
    import seaborn as sns
    import numpy as np
    import scipy
    
    import seaborn as sns
    cols = ['#56AC52','#337848','#F2E92A','#F58320','#822F31','#CC3C39','#475BA6','#78CDDA','#5375BA','#EA3C81'
        ,'#7E441D','#848585','#8E5EA7','#B9D36C','#000000']
    colors = sns.color_palette(cols)
#     sns.palplot(colors)
#     plt.show()
    
    s200=pd.read_csv('./DAB2-Batch5/18Sept27_DABS200/18Sept27_DAB_S200Inrease_Cleaned.csv')
    cv = s200['UV 1_280_ml']/2.4
    plt.plot(cv,s200['UV 1_280_mAU'],label='UV 280', color=colors[6],lw=4)
#     plt.plot(s200['UV 2_485_ml'],s200['UV 2_485_mAU'],label='UV 485')
    plt.xlim(0,max(cv))
    plt.ylabel('Absorbance at 280nm (mAU)',fontsize=26)
    plt.xlabel('Column volumes',fontsize=26)
    plt.xticks(fontsize=26)
    plt.yticks(fontsize=26)
    plt.xlim(0.2,1)
    plt.tick_params(axis='both', direction='out', bottom='on', left='on', top='off', right='off',
                    length=10, width=4)
        # Hide the right and top spines
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)

    # Only show ticks on the left and bottom spines
    ax.yaxis.set_ticks_position('left')
    ax.xaxis.set_ticks_position('bottom')

#     plt.legend()
    
fig,ax = plt.subplots(figsize=(20,10))
makePanelC(ax)
plt.savefig('./../scripts/FigureMakingForBarseqDABPaper/figsForPaper/fig4PanelS3.eps',
           format='eps',bbox_inches='tight')
plt.show()
```
