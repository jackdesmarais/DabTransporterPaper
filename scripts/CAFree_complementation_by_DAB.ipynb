```python
# Making Panel B
%matplotlib inline
def makePanelB(ax):
    import pandas as pd
    import numpy as np
    import seaborn
    import itertools

    from growth.plate_spec import PlateSpec
    from growth.plate_time_course_parser import SavageLabM1000Excel
    from matplotlib import pyplot as plt
    
    import seaborn as sns
    cols = ['#56AC52','#337848','#F2E92A','#F58320','#822F31','#CC3C39','#475BA6','#78CDDA','#5375BA','#EA3C81'
            ,'#7E441D','#848585','#8E5EA7','#B9D36C','#000000']
    colors = sns.color_palette(cols)
    pCol=colors[1]
    nCol=colors[4]
    eCol=colors[6]

    # PlateSpec defined in CSV for easy editing
    ps = PlateSpec.FromFile('growth/plate_specs/2018_02_07_operon1CAFreeKnockInGrowth_metalKO_nice.csv')
    
    parser = SavageLabM1000Excel()
    timecourse = parser.ParseFromFilename('growth/data/2018_02_09_CAFree_operon1_rescue_experiment_metal_KO_later.xlsx')
    blanked = timecourse.blank()
    # You can blank each well against it's own early timepoints
    # by not setting blank_wells

    # Smooth with default params.
    smoothed = blanked.smooth()
    
    
    data = smoothed.data_for_label('abs600')
    
#     print(data)
    to_plot = [
        ('bw+pFE-sfGFP', 'WT vector (+)'),
        ('CAF+pFE-sfGFP', 'vector (-)'),
        ('CAF+pFE-HCA2', 'HCAII (+)'),
        ('CAF+pFE-DAB2_wt', 'DAB2'),
        ('CAF+pFE-NDH', 'DAB2A'),
        ('CAF+pFE-DUF', 'DAB2B'),
        ('CAF+pFE-DAB2+dCys1', 'DAB2 C539A'),
        ('CAF+pFE-DAB2+dHis1', 'DAB2 H539A')]
    
    
    yld = smoothed.GrowthYield(density_label='abs600')
    
#     yld = smoothed.AreaUnderTheCurve(density_label='abs600') #Use this to generate area uer the curve instead of yield
    name2well = ps.name_to_well_mapping()

    yld = pd.DataFrame(yld, index=['yield']).T
    grouped_yld = yld.groupby(ps.well_to_name_mapping())
    mean_yield = grouped_yld.mean()
    std_yield = grouped_yld.std()
    
    labels = [t[1] for t in to_plot]
    ylds = [mean_yield.loc[t[0]][0] for t in to_plot]
    stds = [std_yield.loc[t[0]][0] for t in to_plot]
    
    ax.bar(range(len(ylds)),ylds,yerr=stds,color=[pCol,nCol,pCol,eCol,eCol,eCol,eCol,eCol],align='center',ecolor='k')
    plt.ylabel('Growth Yield (OD600)',fontsize=30)
    plt.xticks(range(len(ylds)),labels,rotation=90,fontsize=30)
    plt.yticks(fontsize=30)
    frame = pd.DataFrame(columns=['Strain','Mean yield','Standard deviation'])
    frame['Strain']=labels
    frame['Mean yield']=ylds
    frame['Standard deviation']=stds
    frame.to_excel('./growth/compiledData/2018_02_09_CAFree_operon1_rescue_experiment_metal_KO_later_averaged.xlsx')

    
fig,ax = plt.subplots(figsize=(10,10))
makePanelB(ax)
plt.show()
```
